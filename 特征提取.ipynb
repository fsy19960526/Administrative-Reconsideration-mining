```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import os
import re
```

```python
df1 = pd.read_excel('行政复议.xlsx',sheet_name='2016')
df2 = pd.read_excel('行政复议.xlsx',sheet_name='2017')
df3 = pd.read_excel('行政复议.xlsx',sheet_name='2018')
df4 = pd.read_excel('行政复议.xlsx',sheet_name='2019')
```

```python
df=pd.concat([df1,df2,df3,df4],axis=0)
```

```python
df
```

```python
#构建法院类别特征
df['court_type']=False
df['court_type'][df['法院'].str.contains('[^中级|最高]人民法院').fillna(False)] ='local'
df['court_type'][df['法院'].str.contains('最高人民法院').fillna(False)]='Supreme Court'
df['court_type'][df['法院'].str.contains('中级人民法院').fillna(False)]='medium'
df['court_type'][df['法院'].str.contains('高级人民法院').fillna(False)]='high'
df['court_type'][df['法院'].str.contains('[^人民]法院').fillna(False)]='normal'
df['court_type']
```

```python
#构建原告类别特征
df['原告'][df['原告'].apply(lambda x:len(str(x))<=3)]#个人
df['原告'][(df['原告'].apply(lambda x:len(str(x).split('、')[0])>3))&(~((df['原告'].str.contains('等').fillna(False))|(df['原告'].apply(lambda x:len(str(x).split('、'))>=2))))]#法人
df['原告'][(df['原告'].str.contains('等').fillna(False))|(df['原告'].apply(lambda x:len(str(x).split('、'))>=2))]#集体
df['plaintiff_type']=False
df['plaintiff_type'][df['原告'].apply(lambda x:len(str(x))<=3)]='personal'
df['plaintiff_type'][(df['原告'].apply(lambda x:len(str(x).split('、')[0])>3))&(~((df['原告'].str.contains('等').fillna(False))|(df['原告'].apply(lambda x:len(str(x).split('、'))>=2))))]='legal entity'
df['plaintiff_type'][(df['原告'].str.contains('等').fillna(False))|(df['原告'].apply(lambda x:len(str(x).split('、'))>=2))]='collective'
df['plaintiff_type']
```

```python
df[df['原告'].apply(lambda x:len(str(x))<=3)][df[df['原告'].apply(lambda x:len(str(x))<=3)]['当事人'].str.contains('汉族').fillna(True)]#个体当事人是否为汉族
```

```python
df['被告'][df['被告'].str.contains('[^公安|分]局').fillna(False)]
```

```python
df['defendant_type']='others'
df['defendant_type'][df['被告'].str.contains('人民政府').fillna(False)] ='gov'
df['defendant_type'][df['被告'].str.contains('公安局').fillna(False)]='public_security_bru'
df['defendant_type'][df['被告'].str.contains('税务局').fillna(False)]='tax_bru'
df['defendant_type'][df['被告'].str.contains('中华人民共和国').fillna(False)]='Central_department'
df['defendant_type'][df['被告'].str.contains('[^公安|分|税务]局').fillna(False)]='other_bru'
df['defendant_type'][df['被告'].str.contains('街道办事处').fillna(False)]='street_office'
df['defendant_type'][df['被告'].str.contains('委员会').fillna(False)]='Committee'
df['defendant_type'][(df['被告'].str.contains('人民政府').fillna(False))&(df['被告'].str.contains('公安局').fillna(False))]='gov_and_public_security_bru'
df['defendant_type'][(df['被告'].str.contains('人民政府').fillna(False))&(df['被告'].apply(lambda x:len(str(x).split('、'))>=2))] ='gov_and_others'
df['defendant_type']
```
